```python

import pandas as pd
import numpy as np


data = {
    'PassengerId': np.arange(1, 891),
    'Survived': np.random.randint(0, 2, size=890),
    'Pclass': np.random.randint(1, 4, size=890),
    'Name': [f"Passenger {i}" for i in range(1, 891)],
    'Sex': np.random.choice(['male', 'female'], size=890),
    'Age': np.random.randint(1, 80, size=890),
    'SibSp': np.random.randint(0, 6, size=890),
    'Parch': np.random.randint(0, 6, size=890),
    'Ticket': [f"Ticket {i}" for i in range(1, 891)],
    'Fare': np.round(np.random.uniform(10, 500, size=890), 2),
    'Embarked': np.random.choice(['C', 'Q', 'S'], size=890)
}

df = pd.DataFrame(data)

print(df.head())

print(df.isnull().sum())


df['Age'].fillna(df['Age'].median(), inplace=True)


df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

print(df.info())


print(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Survived', element='step', stat='density', common_norm=False, palette='pastel')
plt.title('Survival by Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()


plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='pastel')
plt.title('Survival Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8, 5))
sns.histplot(df['Fare'], kde=True, color='coral')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
